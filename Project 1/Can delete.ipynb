```python

import os

import bs4
import matplotlib.pyplot as plt
import pandas as pd
import requests
import sqlalchemy as db
from bs4 import BeautifulSoup
import re
from datetime import datetime
from typing import List
```

```python
TAXI_URL: str = "https://www.nyc.gov/site/tlc/about/tlc-trip-record-data.page"

def get_taxi_html() -> str:
    response = requests.get(TAXI_URL)
    html = response.content
    return html
def find_taxi_parquet_links() -> List[str]:
    ### BEGIN SOLUTION
    html = get_taxi_html()
    soup = bs4.BeautifulSoup(html, "html.parser")
# Find all <a> tags with an href attribute containing ".parquet"
    parquet_links = [
        a_tag["href"] 
        for a_tag in soup.find_all("a", href=True) 
        if ".parquet" in a_tag["href"]
    ]
    return parquet_links

# Example usage
if __name__ == "__main__":
    parquet_links = find_taxi_parquet_links()
    for link in parquet_links:
        print(link)
```
